# Travel Impedance maps (Figure 3)

```
def add_gtc(Ti_df, G):
  """
    Ti_df = Pandas Dataframe with nodes on the index and GTC based Travel Impedance in column
    G = networkx (multi)graph
  This function adds the name of the stops to the nodes in the selected graph
  returns: networkx (multi)graph 
  """
  gtc = dict(zip(Ti_df.index, (Ti_df/60).iloc[:,0]))
  gtc = {int(k):float(v) for k,v in gtc.items()}
  nx.set_node_attributes(G, gtc, "gtc" )

  return G

adelaide89_P_GCC = add_gtc(adelaide_89_travel_impedance, adelaide89_P_GCC)
adelaide1112_P_GCC = add_gtc(adelaide_1112_travel_impedance, adelaide1112_P_GCC)

dublin89_P_GCC = add_gtc(dublin_89_travel_impedance, dublin89_P_GCC)
dublin1112_P_GCC = add_gtc(dublin_1112_travel_impedance, dublin1112_P_GCC)

rome89_P_GCC = add_gtc(rome_89_travel_impedance, rome89_P_GCC)
rome1112_P_GCC = add_gtc(rome_1112_travel_impedance, rome1112_P_GCC)
```

```
def add_topo(Ti_df, G):
  """
    Ti_df = Pandas Dataframe with nodes on the index and GTC based Travel Impedance in column
    G = networkx (multi)graph
  This function adds the name of the stops to the nodes in the selected graph
  returns: networkx (multi)graph 
  """
  topo = dict(zip(Ti_df.index, (Ti_df).iloc[:,0]))
  topo = {int(k):float(v) for k,v in topo.items()}
  nx.set_node_attributes(G, topo, "topo" )
  return G

adelaide89_P_GCC = add_topo(adelaide_89_topo_ti, adelaide89_P_GCC)
adelaide1112_P_GCC = add_topo(adelaide_1112_topo_ti, adelaide1112_P_GCC)

dublin89_P_GCC = add_topo(dublin_89_topo_ti, dublin89_P_GCC)
dublin1112_P_GCC = add_topo(dublin_1112_topo_ti, dublin1112_P_GCC)

rome89_P_GCC = add_topo(rome_89_topo_ti, rome89_P_GCC)
rome1112_P_GCC = add_topo(rome_1112_topo_ti, rome1112_P_GCC)
```

```
def add_gap(Ti_gtc_df, Ti_bench_df, G):
  """
    Ti_gtc_df = Pandas Dataframe with nodes on the index and GTC based Travel Impedance in column
    Ti_bench_df = Pandas Dataframe with nodes on the index and benchmark Travel Impedance in column
    G = networkx (multi)graph
  This function adds the difference gap between the metrics to the nodes in the selected graph
  returns: networkx (multi)graph 
  """
  gap_data = Ti_gtc_df.copy()
  data = pd.merge(Ti_bench_df,(gap_data/60), left_index=True, right_index=True)
  x = data.Ti_x.values
  y = data.Ti_y.values
  x = x.reshape(len(data), 1)
  y = y.reshape(len(data), 1)
  regr = linear_model.LinearRegression()
  regr.fit(x, y)
  y_pred = regr.predict(x)

  gap_data["gap"] = y-y_pred
  gap = dict(zip(gap_data.index, gap_data.iloc[:,1]))
  gap = {int(k):float(v) for k,v in gap.items()}
  nx.set_node_attributes(G, gap, "gap" )
  return G

adelaide89_P_GCC = add_gap(adelaide_89_travel_impedance, adelaide_89_topo_ti, adelaide89_P_GCC)
adelaide1112_P_GCC = add_gap(adelaide_1112_travel_impedance, adelaide_1112_topo_ti, adelaide1112_P_GCC)

dublin89_P_GCC = add_gap(dublin_89_travel_impedance, dublin_89_topo_ti, dublin89_P_GCC)
dublin1112_P_GCC = add_gap(dublin_1112_travel_impedance, dublin_1112_topo_ti, dublin1112_P_GCC)

rome89_P_GCC = add_gap(rome_89_travel_impedance, rome_89_topo_ti, rome89_P_GCC)
rome1112_P_GCC = add_gap(rome_1112_travel_impedance, rome_1112_topo_ti, rome1112_P_GCC)
```

```
def count_stops_per_mode(P_space):
  """
  This code computes a list per mode of transport with corresponding stops and 
    a pd.dataframe with the count per mode
  Input: P-space Edge list
  Returns: pd.DataFrame with count per mode of transport and a list per mode of transport
  """
  count_tram = 0
  list_tram = []
  count_subway = 0
  list_subway = []
  count_rail = 0
  list_rail = []
  count_bus = 0
  list_bus = []
  for row in P_space.itertuples():
    if row[8] == 0:
      if row[1] not in list_tram:
        list_tram.append(row[1])
        count_tram += 1
      if row[2] not in list_tram:
        list_tram.append(row[2])
        count_tram += 1
    elif row[8] == 1:
      if row[1] not in list_subway:
        list_subway.append(row[1])
        count_subway += 1
      if row[2] not in list_subway:
        list_subway.append(row[2])
        count_subway += 1
    elif row[8] == 2:
      if row[1] not in list_rail:
        list_rail.append(row[1])
        count_rail += 1
      if row[2] not in list_rail:
        list_rail.append(row[2])
        count_rail += 1
    elif row[8] == 3:
      if row[1] not in list_bus:
        list_bus.append(row[1])
        count_bus += 1
      if row[2] not in list_bus:
        list_bus.append(row[2])
        count_bus += 1
  count_df = pd.DataFrame ({"tram_nodes": count_tram, "subway_nodes": count_subway,
                            "rail_nodes": count_rail, "bus_nodes": count_bus}, 
                           index = ["count"])
  return count_df, list_tram, list_subway, list_rail, list_bus

```

```
def statistics_per_mode_rome(matrix_ti, list_tram, list_rail, list_subway, list_bus):
  """
  This code computes a pd.DataFrame comprising statistics per mode of transport in the Rome networks
  Rome has a separate functions since Rome's PTN comprises subway, which the others don't have
  Input: matrix_ti = gtc-based travel impedance per stop, lists with nodes per mode
  of transport
  Returns: pd.DataFrame with count per mode of transport and a list per mode of transport
  """
  matrix_ti = (matrix_ti/60)
  tram_dict = {} #empty dictionary for gtc based travel impedance 
  rail_dict = {} #empty dictionary for gtc based travel impedance
  subway_dict = {} #empty dictionary for gtc based travel impedance 
  bus_dict = {} #empty dictionary for gtc based travel impedance  
  mode_col = ["total", "tram", "rail", "subway", "bus"]
  count_col = []
  mean_col = []
  std_col = []
  count_col.append(len(matrix_ti)) #n of stops total
  mean_col.append(matrix_ti["Ti"].mean()) #mean of gtc based travel impedance total
  std_col.append(matrix_ti["Ti"].std()) #std of gtc based travel impedance total
  for i in list_tram:
    for row in matrix_ti.itertuples():
      if i == int(row[0]):
        tram_dict[i] = row[1]
  tram_df = pd.DataFrame.from_dict(tram_dict, orient = "index")
  count_col.append(len(tram_df)) #n of stops tram
  mean_col.append(tram_df[0].mean()) #mean of gtc based travel impedance tram
  std_col.append(tram_df[0].std()) #std of gtc based travel impedance tram

  for i in list_rail:
    for row in matrix_ti.itertuples():
      if i == int(row[0]):
        rail_dict[i] = row[1]
  rail_df = pd.DataFrame.from_dict(rail_dict, orient = "index") #dict to pd.DataFrame
  count_col.append(len(rail_df)) #n of stops rail
  mean_col.append(rail_df[0].mean()) #mean of gtc based travel impedance rail
  std_col.append(rail_df[0].std()) #std of gtc based travel impedance rail

  for i in list_subway:
    for row in matrix_ti.itertuples():
      if i == int(row[0]):
        subway_dict[i] = row[1]
  subway_df = pd.DataFrame.from_dict(subway_dict, orient = "index") #dict to pd.DataFrame
  count_col.append(len(subway_df)) #n of stops subway
  mean_col.append(subway_df[0].mean()) #mean of gtc based travel impedance subway
  std_col.append(subway_df[0].std()) #std of gtc based travel impedance subway

  for i in list_bus:
    for row in matrix_ti.itertuples():
      if i == int(row[0]):
        bus_dict[i] = row[1]
  bus_df = pd.DataFrame.from_dict(bus_dict, orient = "index") #dict to pd.DataFrame
  count_col.append(len(bus_df)) #n of stops bus
  mean_col.append(bus_df[0].mean()) #mean of gtc based travel impedance bus
  std_col.append(bus_df[0].std()) #std of gtc based travel impedance bus

  data_tuples = list(zip(mode_col, count_col, mean_col, std_col)) #put lists into tuples
  statistics = pd.DataFrame(data_tuples, columns=['mode','count', "mean", "std"]) #convert tuples to df
  return statistics
  
```

```
def create_graph_ti_difference(city_ti_difference_df, network_node_list):
  """
  This code creates two nx.Graphs. The first one of nodes that appear in both timeframes,
    with the coordinates of the nodes and the difference in gtc-based travel 
    impedance between rush hour and off-peak hours as node attributes.
    The second one of nodes that appear in only one timeframe and, therefore, lack
    a "travel impedance difference" (i.e. =nan)
    These graphs can be used plot the difference in travel impedance in a map.
  Input: city_ti_difference_df = pd.DataFrame consisting of the travel impedance 
    per stop in the different timespans. Third column has to exist of the difference in travel time
    network_node_list = pd.DataFrame comprising the nodes and corresponding coordinates in latitude and longitude values
  Returns: G = nx.Graph consisting of the nodes (and node attributes) that appear in both the 8-9 and 11-12 network
    G_nan = nx.Graph consisting of the nodes that appear in only one of the networks
  """
  G = nx.Graph() #create empty nx.Graph for the nodes with ti_difference value
  G_nan = nx.Graph() #create empty nx.Graph for the nodes without ti_difference value(=nan)

  city_ti_difference_good = city_ti_difference_df[city_ti_difference_df["ti_difference"].notnull()]
  city_ti_difference_nan =  city_ti_difference_df[city_ti_difference_df["ti_difference"].isnull()]
 
  G.add_nodes_from(city_ti_difference_good.index.values.astype(int)) #add nodes with value to G by using index of ti_difference matrix
  network_node_list['coord'] = list(zip(network_node_list.lon, network_node_list.lat)) #create column with coordinates in network_node_list
  pos = network_node_list.set_index("stop_I")["coord"].to_dict() #create dictionary pos, comprising the coordinates
  pos = {int(k):v for k,v in pos.items()} #convert keys to integer
  nx.set_node_attributes(G, pos, name = "pos") #add coordinates to nodes G

  ti_diff_good = dict(zip(city_ti_difference_good.index, (city_ti_difference_good).iloc[:,2])) #create dictionary 
  #ti_diff_good, comprising the travel impedance difference per stop in G
  ti_diff_good = {int(k):float(v) for k,v in ti_diff_good.items()} #convert keys to integer
  nx.set_node_attributes(G, ti_diff_good, "ti_diff" ) #add the difference in travel impedance as node attritube to G

  G_nan.add_nodes_from(city_ti_difference_nan.index.values.astype(int)) #add nodes to graph by using index of ti_difference matrix
  nx.set_node_attributes(G_nan, pos, name = "pos") #add coordinates to nodes in nx.Graph

  return G, G_nan

adelaide_diff, adelaide_diff_nan = create_graph_ti_difference(adelaide_ti_difference, nodes_adelaide)
dublin_diff, dublin_diff_nan = create_graph_ti_difference(dublin_ti_difference, nodes_dublin)
rome_diff, rome_diff_nan = create_graph_ti_difference(rome_ti_difference, nodes_rome)
```
